```python
%load_ext autoreload
%autoreload 2

import torch
import numpy as np
from torch_nf.conditional_nf import NormFlow, ConditionedNormFlow
import scipy.stats
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
D = 4
eps = 1e-5
bn = torch.nn.BatchNorm1d(D, eps=eps, momentum=.9, affine=False)
```

```python
for i in range(1000):
    
    z = np.random.normal(10., 2., (100, 4))
    z = torch.tensor(z).float()
    z_mean = torch.mean(z, axis=0)
    z_var = torch.var(z, axis=0, unbiased=False)
    z_out = bn(z)
    
    error = torch.sum((((z-z_mean)/np.sqrt(z_var+eps))-z_out)**2)

    bn_mean = bn.running_mean[None, :]
    bn_var = bn.running_var[None, :]
    est_error = torch.sum((((z-bn_mean)/np.sqrt(bn_var+eps))-z_out)**2)
    
    print(i, error, est_error)
```
